#Classifying foods in order to fulfill dietary resctricitons, in this case Diabetes, Low-sodium and Cholesterol-restricted diets

```python
print('Average carbohydrates value:', food_data_filtered['Carbohydrates'].mean())
print('Average cholesterol value:', food_data_filtered['Cholesterol'].mean())
print('Average sodium value:', food_data_filtered['Sodium'].mean())
```

```python
#Setting thresholds for each dietary resctriction
carbohydrates_threshold = 15
cholesterol_threshold = 0.1
sodium_threshold = 0.2
```

```python
#Creating columns for every food considering every set restriction
food_data_filtered['Low_carbohydrates'] = np.where(food_data_filtered['Carbohydrates'] <= carbohydrates_threshold, True, False)
food_data_filtered['Low_sodium'] = np.where(food_data_filtered['Sodium'] <= sodium_threshold, True, False)
food_data_filtered['Low_cholesterol'] = np.where(food_data_filtered['Cholesterol'] <= cholesterol_threshold, True, False)
```

```python
#Importing standard scaler and scaling data to achieve equality in values of different columns
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
#Scaling selected features
features = ['Caloric Value', 'Carbohydrates', 'Cholesterol', 'Sodium']
data_scaled = scaler.fit_transform(food_data_filtered[features])
```

```python
#Checking the scaled features
pd.DataFrame(data_scaled, columns=features)
```

```python
pd.DataFrame(data_scaled, columns=features).describe()
```

```python
#Combining scaled data with food, low_carbohydrates, low_sodium, low_cholesterol columns
scaled_df = pd.DataFrame(data_scaled, columns=features)
scaled_df = pd.concat([food_data_filtered[['food', 'Low_carbohydrates', 'Low_sodium', 'Low_cholesterol']], scaled_df], axis=1)
scaled_df
```

#Recommendation system

```python
#Importing euclidean distances
from sklearn.metrics.pairwise import euclidean_distances
```

```python
#Creating pairs based on euclidian distances
euclidean_matrix = euclidean_distances(data_scaled)
euclidean_matrix.shape
```

```python
scaled_df
```

```python
def recommender_food(food, n_recs):
    if food in list(set(scaled_df.food)):
        ix = scaled_df[scaled_df['food'] == food].index[0]
        ixs = np.argsort(euclidean_matrix[ix, :])[1:n_recs+1]
        return scaled_df['food'][ixs]
    else:
        return 'Food not in the dataset'
```

```python
def recommender_food2(food, n_recs):
    if str(food) in scaled_df['food'].astype(str).unique():
        #Find the index of the food
        ix = scaled_df[scaled_df['food'] == food].index[0]

        #Find the nearest neighbors' indices
        ixs = np.argsort(euclidean_matrix[ix, :])[1:n_recs+1]

        #Return the recommended foods
        return scaled_df.iloc[ixs]['food'].tolist()
    else:
        return 'Food not in the dataset'
```

```python
recommender_food('cream cheese', 5)
```

#Creating a recommendation function taking into consideration dietary restrictions

```python
recommended_foods_list = []

def recommender_food_restrictions(food, n_recs, restrictions=None):
    #Checking if the food exists in the dataset
    if food in list(scaled_df['food']):
        #Indexing of the selected food
        ix = scaled_df[scaled_df['food'] == food].index[0]

        #Filtering dataset based on restrictions, if provided
        if restrictions:
            # Applying all specified restrictions
            valid_foods = scaled_df
            for restriction in restrictions:
                if restriction in ['Low_sodium', 'Low_cholesterol', 'Low_carbohydrates']:
                    valid_foods = valid_foods[valid_foods[restriction] == True]
        else:
            valid_foods = scaled_df

        #Finding the indices of the valid foods in the original DataFrame
        valid_indices = valid_foods.index.tolist()

        #Computing distances only for valid foods
        distances = [(i, euclidean_matrix[ix, i]) for i in valid_indices if i != ix]
        distances = sorted(distances, key=lambda x: x[1])[:n_recs]

        #Returning the names of the recommended foods
        recommended_foods_list = [scaled_df.iloc[i]['food'] for i, _ in distances]
        return recommended_foods_list
    else:
        return 'Food not in the dataset'
```

```python
recommender_food_restrictions('neufchatel cheese', 10, restrictions=['Low_cholesterol'])
```

```python
#Generating meal ideas using Ollama
def generate_meal_ideas_with_ollama(ingredients, num_meals=5):
    ingredient_list = ", ".join(ingredients)
    prompt = (
        f"I have the following ingredients: {ingredient_list}. "
        f"Please generate {num_meals} creative and delicious meal ideas using these ingredients. "
        f"Provide the meal ideas as a numbered list."
    )

    #Making a POST request to the Ollama API
    try:
        url = "http://0.0.0.0:11434/api/generate"
        payload = {"model": "qwen2.5", "prompt": prompt}  # Adjust model name if needed
        response = requests.post(url, json=payload)

        if response.status_code == 200:
            #Parsing the response content
            return response.json().get("text", "No response text found").strip()
        else:
            return f"Error: {response.status_code}, {response.text}"
    except Exception as e:
        return f"An error occurred: {e}"

#Example usage
ingredients = recommended_foods_list
meal_ideas = generate_meal_ideas_with_ollama(ingredients, num_meals=5)
print(meal_ideas)
```
