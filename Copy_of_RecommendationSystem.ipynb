```python
import pandas as pd
import numpy as np
import gradio as gr
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#Create Recommendation Model
def build_content_based_model(items_data):
    # Using TF-IDF for content-based similarity on item descriptions
    vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = vectorizer.fit_transform(items_data['description'])
    similarity_matrix = cosine_similarity(tfidf_matrix)
    return similarity_matrix
```

```python
def recommend_items(user_id, num_recommendations=5):
    # Filter user interactions
    user_interactions = interactions_data[interactions_data['user_id'] == user_id]
    interacted_items = user_interactions['item_id'].tolist()

    # Build a similarity matrix
    similarity_matrix = build_content_based_model(items_data)

    # Aggregate scores for items the user interacted with
    scores = np.zeros(similarity_matrix.shape[0])
    for item_id in interacted_items:
        item_index = items_data[items_data['item_id'] == item_id].index[0]
        scores += similarity_matrix[item_index]

    # Exclude items already interacted with
    scores = [(i, score) for i, score in enumerate(scores) if items_data.iloc[i]['item_id'] not in interacted_items]
    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    # Get top recommendations
    recommendations = [items_data.iloc[i]['title'] for i, _ in scores[:num_recommendations]]
    return recommendations
```

```python
import pandas as pd
import numpy as np
import gradio as gr
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

# Assuming you have your data loading and model building logic here...

#Create Gradio App
def recommend(user_id):
    try:
        user_id = int(user_id)
        recommendations = recommend_items(user_id)
        return "\n".join(recommendations)
    except Exception as e:
        return f"Error: {str(e)}"

app = gr.Interface(
    fn=recommend,
    inputs=gr.Textbox(lines=1, placeholder="Enter your User ID here..."), # Changed to gr.Textbox
    outputs=gr.Textbox(label="Recommended Items"), # Changed to gr.Textbox
    title="Recommendation System",
    description="Enter a user ID to get recommended items based on their preferences.",
)

if __name__ == "__main__":
    app.launch()
```
